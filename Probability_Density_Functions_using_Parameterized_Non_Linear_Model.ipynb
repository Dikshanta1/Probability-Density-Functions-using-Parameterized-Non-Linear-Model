```python
import pandas as pd
import numpy as np
import math
```

```python
data = pd.read_csv('/content/drive/MyDrive/mydata.csv', encoding='latin1', low_memory=False);
```

```python
print(data.head());
```

```python
no2vals = data["no2"].dropna().values
```

```python
print("Total values:", len(no2vals))
```

```python
ar = 0.20
br = 0.60
```

```python
zvals = []
for i in no2vals:
  clipped_arg = np.clip(br * i, -1.0, 1.0)
  value = i + ar * math.asin(clipped_arg)
  zvals.append(value)

zarray = np.array(zvals)

print("now i think transformation in completed")
```

```python
mu = np.mean(zarray)
variance = np.var(zarray)

lambdaValue = 1 / (2 * variance)
cValue = math.sqrt(lambdaValue / math.pi)

print("lambda is", lambdaValue)
print("mu is", mu)
print("c is", cValue)
```
